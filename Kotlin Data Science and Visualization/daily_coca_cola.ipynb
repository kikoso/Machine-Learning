```kotlin
%useLatestDescriptors
%use dataframe
%use kandy
```

```kotlin
val dfRaw = DataFrame.readCSV("data/daily_KO.csv")
dfRaw.head()

```

```kotlin
dfRaw.describe()
```

```kotlin
dfRaw.plot {
    line {
        x(timestamp)
        y(close)
    }
}
```

```kotlin
dfRaw.plot {
    line {
        x(timestamp)
        y(close)
        width = 2.0
        color = Color.RED
    }
    layout {
        size = 800 to 400
        title = "KO Daily Price"
    }
}
```

```kotlin
dfRaw.plot {
    candlestick(timestamp, open, high, low, close)
}
```

```kotlin
dfRaw.plot {
    ribbon {
        x(timestamp)
        y {
            axis.name = "Cost"
            scale = continuous(60.0..71.0)
        }
        yMin(low)
        yMax(high)
        fillColor = Color.hex(0x3f21e6)
        alpha = 0.65
        borderLine {
            color = Color.RED
            width = 0.8
            type = LineType.DASHED
        }
    }
}
```
